```python
df=pd.read_csv('/kaggle/input/remote-work-productivity/remote_work_productivity.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Employment_Type',data=df,palette='GnBu')
plt.title('Amount of Office Workers vs Remote')
plt.xlabel('Employment Type')
plt.show()
```

```python
dftype=df['Employment_Type']

count=dftype.value_counts()
plt.figure(figsize=(10,6))
plt.pie(count,labels=count.index,autopct='%1.1f%%', startangle=90)
plt.title('Proportions of Office Wokers Vs. Remote')
plt.show()
```

Around the same office workers and remote workers surveyed based on the pie chart and bar plot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Employment_Type',y='Hours_Worked_Per_Week',data=df,palette='GnBu')
plt.title('Hours Worked Per Week by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Hours Worked Per Week')
plt.show()
```

The barplot for remote workers is lower than the office barplot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Employment_Type',y='Productivity_Score',data=df)
plt.title('Productivity Score by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Productivity Score')
plt.show()
```

The remote worker bar plot is higher than the in-office bar plot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Employment_Type',y='Well_Being_Score',data=df,palette='Paired')
plt.title('Well Being Score by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Well Being Score')
plt.show()
```

The remote bar plot is also higher than the in-office bar plot here.

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Employment_Type', y='Productivity_Score', data=df,palette='Greens_d')
plt.title('Average Productivity by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Productivity Score')
plt.show()
```

The average productivity for remote workers is much higher compared to office workers.

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Employment_Type', y='Well_Being_Score', data=df,palette='Greens_d')
plt.title('Average Well Being Score by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Well Being Score')
plt.show()
```

The well being score is also much higher for remote workers.

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Employment_Type', y='Hours_Worked_Per_Week', data=df,palette='Greens_d')
plt.title('Average Hours Worked Per Week by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Hours')
plt.show()
```

Office workers work much more hours than remote workers on average.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Hours_Worked_Per_Week'],df['Productivity_Score'])
plt.title('Producitvity Based on Hours Worked Per Week')
plt.xlabel('Productivity Score')
plt.ylabel('Hours')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Hours_Worked_Per_Week'],df['Well_Being_Score'])
plt.title('Well Being Based on Hours Worked Per Week')
plt.xlabel('Well Being Score')
plt.ylabel('Hours')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Well_Being_Score'],df['Productivity_Score'])
plt.title('Producitvity Based on Well Being')
plt.xlabel('Well Being Score')
plt.ylabel('Productivity Score')

```

There isn't a strong correlation in any three of the plots, most likely because the remote and office factors affect their responses.

```python
number = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,8))
sns.heatmap(number.corr(),annot=True, fmt='.2f')
```

There isn't a strong correlation between any of the categories, again for the reason stated in the last markdown.

**Conclusions**

-Office workers typically work much more hours than remote workers.

-Remote workers have higher productivity and well-being compared to office workers.

-There aren't strong correlations between any numerical variables (Well Being, Productivity, Hours), mainly due to the Remote and Office factors affecting the responses.






**If remote workers work less hours than office workers, on average, and have better productivity and well being, then that may be a sign that remote work is much more efficient than office work.**

**One limitation is that we don't know how productivity is measured, whether by someone else looking over or the surveyor.**
